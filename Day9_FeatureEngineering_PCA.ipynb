Feature engineering means creating new meaning features from existing data to improve ML performance.

Used in :


*  EV battery SOH detection

*  Renewable hostpot mapping

*  fault detection
*  Image processing


*   Stock prediction






```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
data = load_iris()
x = data.data
y = data.target
df = pd.DataFrame(x, columns = data.feature_names)
df['class'] = y
df.head()
```

```python
df['petal ratio'] = df['petal length (cm)']/df['petal width (cm)']
```

```python
df['sepal diff']=df['sepal length (cm)'] - df['sepal width (cm)']
```

```python
df['interaction']=df['sepal length (cm)']*df['petal width (cm)']
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler #scaling data

features = df.drop('class', axis=1)
scaler = StandardScaler()
x_scaled = scaler.fit_transform(features)
```

```python
from sklearn.decomposition import PCA

pca=PCA(n_components = 2)
x_pca=pca.fit_transform(x_scaled)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1], c=y, cmap='viridis')
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.title("PCA on engineered features(Iris)")
plt.show()
```
