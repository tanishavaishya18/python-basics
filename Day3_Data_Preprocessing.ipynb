```python
import pandas as pd
import numpy as np

data = {
    'age': [20, 25, np.nan, 30, 22],
    'income': [50000, np.nan, 45000, 60000, 52000],
    'city': ['Mumbai', 'Delhi', 'Chennai', np.nan, 'Pune']
}

df = pd.DataFrame(data)
df
```

```python
df.isnull().sum() #counts missing values
df.dropna() #removes rows with any missimg value

df['age'] = df['age'].fillna(df['age'].mean()) #fills the unknown values with the mean of the column
df['city']=df['city'].fillna('Unknown') #fills the missing city values with unknown

```

```python
df
```

```python
df['age_scaled'] = (df['age'] - df['age'].mean()) / df['age'].std() #standardisation
#used for getting the data in same scale and making teh mean of the output data 0 and std deviation 1
```

```python
df
```

```python
df['income'] = df['income'].fillna(df['income'].mean())
df
```

```python
df['income_scaled'] = (df['income'] - df['income'].min()) / (df['income'].max() - df['income'].min()) #mign-max scalin
df
```
