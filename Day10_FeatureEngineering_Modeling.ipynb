```python
# USING ENGINEERED FEATURES FOR ML
```

Why engineeered featutres improve accuracy:



*   better signal extraction

*   higher separability between classes
*   captures non-linear relationships


*   reduces model confusion



In this notebook I have -





*  Trained Logistic Regression
*  Trained random forest


*  Compared the results before vs after feature engineering




```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
data = load_iris()

x=data.data
y=data.target
df = pd.DataFrame(x, columns=data.feature_names)
df['class']=y
```

```python
from sklearn.model_selection import train_test_split #train test split

x_orig = df.drop('class', axis=1)
y=df['class']

x_train, x_test, y_train, y_test = train_test_split(
    x,y,test_size =0.2, random_state=42
)
```

```python
#training logistic regression

from sklearn.linear_model import LogisticRegression

log_reg_orig = LogisticRegression(max_iter = 200)
log_reg_orig.fit(x_train, y_train)
orig_pred = log_reg_orig.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

baseline_accuracy = accuracy_score(y_test, orig_pred)
baseline_accuracy
```

```python
df['petal_ratio'] = df['petal length (cm)'] / df['petal width (cm)']
df['sepal_diff']  = df['sepal length (cm)'] - df['sepal width (cm)']
df['interaction'] = df['sepal length (cm)'] * df['petal width (cm)']
```

```python
x_eng = df.drop('class', axis=1)
x_train_eng, x_test_eng, y_train, y_test = train_test_split(
    x_eng, y, test_size=0.2, random_state=42
)

log_reg_eng = LogisticRegression(max_iter = 200)
log_reg_eng.fit(x_train_eng, y_train)
eng_pred = log_reg_eng.predict(x_test_eng)
eng_accuracy = accuracy_score(y_test, eng_pred)
eng_accuracy
```

```python
# RANDOM FOREST COMPARISON
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_orig = RandomForestClassifier(random_state = 42)
rf_orig.fit(x_train, y_train)
pred_orig = rf_orig.predict(x_test)
rf_orig_accuracy = accuracy_score(y_test, pred_orig)
rf_orig_accuracy
```

```python
X_eng = df.drop('class', axis=1)

X_train_eng, X_test_eng, y_train, y_test = train_test_split(
    X_eng, y, test_size=0.2, random_state=42
)

log_reg_eng = LogisticRegression(max_iter=200)
log_reg_eng.fit(X_train_eng, y_train)
eng_pred = log_reg_eng.predict(X_test_eng)
eng_accuracy = accuracy_score(y_test, eng_pred)
eng_accuracy
```

We see an improvement in the accurcay score after feature engineering, here both the times it is 1 because the initial dataframe was already perfectly separable.
